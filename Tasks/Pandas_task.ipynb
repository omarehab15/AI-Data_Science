```python
import pandas as pd

data = {
    'Name': ['Omar', 'Sara', 'Ali', 'Mona', 'Ahmed'],
    'Age': [23, 29, 21, 24, 30],
    'Score': [90, 85, 70, 95, 88]
}

df = pd.DataFrame(data)
average_age = df['Age'].mean()
max_score = df['Score'].max()
high_score = df[df['Score'] > 80]['Name'].tolist()

df


```

```python
print(f'The average age is: {average_age}')
print(f'The maximum score is: {max_score}')
print(f'The names of students who got more than 80: {high_score}')
```

```python
data = {
    'Name': ['Omar', 'Sara', 'Ali', 'Mona', None, 'Ahmed'],
    'Age': [23, None, 21, 24, 27, 30],
    'Score': [90, 85, None, 95, 75, 88]
}

df = pd.DataFrame(data)

df
```

```python
df.info()
```

```python
age_mean = df['Age'].mean()

age_mean
```

```python
df['Age'].fillna(age_mean, inplace=True)
```

```python
df['Score'].fillna(df['Score'].min(), inplace=True)
df
```

```python
df.dropna(subset=['Name'], inplace=True)
df
```

```python
data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Mouse'],
    'Price': [15000, 300, 700, 5000, 350],
    'Quantity': [2, 5, 3, 1, 4]
}

df = pd.DataFrame(data)
df
```

```python
# add a new column
df['Total'] = df['Price'] * df['Quantity']
df
```

```python
total_revenue = df['Total'].sum()
most_sold_product = df.loc[df['Quantity'].idxmax(), 'Product']
```

```python
print(f'The total revenue is: {total_revenue}')
print(f'The most sold product is: {most_sold_product}')
df
```
