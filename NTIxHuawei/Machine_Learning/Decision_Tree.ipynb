```python
import numpy  as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import  load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
data=load_iris()
dataset=data.data
Label=data.target
Label
```

```python
X_train, X_test, y_train, y_test=train_test_split(dataset,Label ,test_size=0.2,random_state=123)
```

```python
model = DecisionTreeClassifier(criterion='gini',max_depth=3)
model.fit(X_train, y_train)  #Training stage
```

```python
print(model.predict(X_test) )       #Prediction  , eval
print(y_test)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
# save the model to disk
import pickle
filename = 'DecisionTree_93.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
import graphviz
from sklearn   import  tree
dot_data = tree.export_graphviz(model, out_file=None,
                               feature_names=list(data.feature_names),
                               class_names=list(data.target_names),filled=True,
                                rounded=True,
                              special_characters=True)
graph = graphviz.Source(dot_data)
graph.render("Gini")
graph
```
