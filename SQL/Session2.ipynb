```python
from sqlalchemy import text
#Define your SQL query
sql_query = text("SHOW TABLES")

#Execute the query and fetch results 
with engine.connect() as conn:
    result = conn.execute(sql_query)
    for row in result:
        print(row)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection details
username = 'root'
password = 'root'
host = 'localhost'
database = 'debi'

# Create connection string and engine
connection_string = f'mysql+mysqlconnector://{username}:{password}@{host}/{database}'
engine = create_engine(connection_string)

# Read CSV file into a DataFrame
df = pd.read_csv('data/mobile_prices.csv')

# Write DataFrame to MySQL table
df.to_sql('mobile', con=engine, if_exists='replace', index=False)

# Verify data insertion
result_df = pd.read_sql('SELECT * FROM mobile', con=engine)
print(result_df.columns)
```

```python
from mysql import connector
mydb = connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'root',
    database = 'debi'
)
cursor = mydb.cursor()
cursor.execute("SHOW TABLES")
cursor.fetchall()
```

```python
cursor.execute("SELECT * FROM mobile WHERE battery_power BETWEEN 1000 AND 2000")
cursor.fetchall()
```

```python
cursor.execute("SELECT battery_power FROM mobile WHERE ram > 3000 AND talk_time >10")
cursor.fetchall()
```

```python
cursor.execute("SELECT battery_power/100 FROM mobile WHERE battery_power > 1000")
cursor.fetchall()
```

```python
cursor.execute("SELECT battery_power FROM mobile ORDER BY battery_power DESC")
cursor.fetchall()
```

```python
query = "INSERT INTO employee (Name,ID,DOB,Address) VALUES (%s,%s,%s,%s)"
values = [
    ('Islam','123','1993-03-25','6 October'),
    ('Ahmed','125','1997-03-25','Giza'),
    ('Hany','126','1998-03-25','Cairo')
]

cursor.executemany(query,values)
```

```python
cursor.execute("CREATE TABLE salary (ID VARCHAR(20), Salary VARCHAR(20))")
```

```python
query = "INSERT INTO salary (ID,Salary) VALUES (%s,%s)"
values = [
    ('123','5000'),
    ('125','6000'),
    ('126','7000')
]

cursor.executemany(query,values)
```

Inner Join

```python
cursor.execute("SELECT employee.Name, salary.Salary FROM salary INNER JOIN employee ON employee.ID=salary.ID")
cursor.fetchall()
```
